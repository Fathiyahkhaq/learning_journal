# Case Study Cosine Similarity

```python
#case study kesamaan item
item1 = np.array([0,0,0,1,0,1,0,0,0,1,0,0,0,1,0,0,0,1,0,0])
item2 = np.array([0,0,0,1,0,1,0,0,0,1,0,0,0,1,1,0,0,0,1,0])
dot = item1 @ item2
cos = dot/(np.linalg.norm(item1)*np.linalg.norm(item2))

print("Hasil dot product :", dot)
print("cosine similarity :", cos)
```

tingkat kemiripan dari item 1 dan 2 adalah 73% sehingga dapat disimpulkan bahwa item 1 dan 2 adalah hampir mirip

# Luas Latvia

```python
import pandas as pd
import numpy as np

data_geo_map = pd.read_csv('https://public.opendatasoft.com/explore/dataset/world-administrative-boundaries/download/?format=csv&timezone=Asia/Jakarta&lang=en&use_labels_for_header=true&csv_separator=%3B',delimiter=';')
data_geo_map.head()
```

```python
Latvia = data_geo_map[data_geo_map['English Name']=='Latvia']['Geo Shape'].iloc[0]
type(Latvia)
```

```python
data_geo_map[data_geo_map['English Name']=='Latvia']['Geo Shape'].shape
```

```python
from ast import literal_eval

latvia_coor_dict = literal_eval(Latvia)

latvia_coor = np.array(latvia_coor_dict['coordinates'])
```

```python
type(latvia_coor_dict)
```

```python
latvia_coor
```

```python
latvia_coor = latvia_coor.reshape(latvia_coor.shape[1],latvia_coor.shape[2]).T

latvia_coor = pd.DataFrame({'long':latvia_coor[0],'lat':latvia_coor[1]})
latvia_coor.head()
```

```python
import matplotlib.pyplot as plt

plt.plot(latvia_coor.long,latvia_coor.lat)
```

```python
median = latvia_coor.lat.median()
median
```

```python
from scipy.integrate import trapz

latvia_coor_sorted = latvia_coor.sort_values('long',ascending=True)

northern = latvia_coor_sorted[latvia_coor_sorted.lat >= median]
southern = latvia_coor_sorted[latvia_coor_sorted.lat < median]

area_N = trapz(northern.lat,northern.long) #trapz(Y,X)
area_S = trapz(southern.lat,southern.long)

print('Luas area Latvia:',np.abs(area_N - area_S))
```
