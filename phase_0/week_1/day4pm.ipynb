## Query Dataset

pengen ambil kondisi fran id=celtics,
                     pts>100,
                     year = 2001
                     lokasinya Home
                     hasilnya Win

```python
df[
    (df['year_id']==2001) &
    (df['fran_id']=='Celtics') &
    (df['game_location']=='H') &
    (df['game_result']=='W')
][['opp_fran','opp_pts','fran_id','pts']]
```

```python
df.query(
    "year_id==2001 & \
    fran_id=='Celtics' & \
    game_location=='H' & \
    game_result=='W'"
    )[['opp_fran','opp_pts','fran_id','pts']]
```

```python
df.groupby('game_result')[['pts','opp_pts']].mean()
```

```python
df[
    (df['year_id']>2010)&
    (df['fran_id']=='Spurs')
].groupby(['year_id','game_result'])['game_id'].count()
```

```python
df[
    (df['year_id']>2010)&
    (df['fran_id']=='Spurs')
].groupby(['year_id','game_result']).agg({'pts':'mean', 'game_id': 'count'})
```
