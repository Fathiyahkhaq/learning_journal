# 5 - Data preprocessing

```python
#get data for model inference
data_inf = data.sample(10,random_state=17)
data_inf
```

## Feature Selection


```python
#display trainset
X_train
```

Karena 'name' gada hubungan dengan 'Rating' maka bisa diasumsikan bahwa kolom ini dihilangkan. hal ini dibuktikan nama 'Danu Ronaldo' tidak ada kaitannya dengan 'Cristiano Ronaldo' sehingga tdk ada pengaruh apapun dg 'Rating'

```python
# Drop column 'Name'
X_train.drop('Name',axis=1,inplace=True)
X_test.drop('Name',axis=1,inplace=True)
X_train
```

```python
X_test
```

# Split Numerical Column and Categorical Column

```python
num1_columns = data.select_dtypes(include=np.number).columns.tolist()
```

```python
print('num : ',num1_columns)
```

```python
num_columns = X_train.select_dtypes(include=np.number).columns.tolist()
cat_columns = X_train.select_dtypes(include=['object']).columns.tolist()
print('num : ',num_columns)
print('cat : ',cat_columns)
```

```python
X_train_num = X_train[num_columns]
X_train_cat = X_train[cat_columns]

X_test_num = X_test[num_columns]
X_test_cat = X_test[cat_columns]
```

```python
## Feature Scalling
X_train_num
```

```python
X_test_num
```

```python
X_train_cat
```

```python
X_test_cat
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(X_train_num) #fit hanya boleh di train gaboleh di test, supaya ga terjadi kebocoran data

X_train_num_scaled = scaler.transform(X_train_num)
X_test_num_scaled = scaler.transform(X_test_num)
X_train_num_scaled
```

```python
X_test_num_scaled.shape
```

## Feature Encoding

```python
from sklearn.preprocessing import OrdinalEncoder #label encoder/one hot encoder
encoder = OrdinalEncoder()
encoder.fit(X_train_cat)

X_train_cat_enc = encoder.transform(X_train_cat)
X_test_cat_enc = encoder.transform(X_test_cat)
X_train_cat_enc
```

```python
X_test_cat_enc
```

```python
#concat num dan cat columns
X_train_final = np.concatenate([X_train_num_scaled, X_train_cat_enc],axis=1)
X_test_final = np.concatenate([X_test_num_scaled, X_test_cat_enc],axis=1)
```

```python
X_train_final
```

```python
X_train_final_df = pd.DataFrame(X_train_final,columns=[num_columns + cat_columns])
X_train_final_df
```

```python
X_test_final
```

```python
X_train_final.shape
```

```python
X_test_final.shape
```

# 6 - Model Definition

```python
# Pake Linear Regression
from sklearn.linear_model import LinearRegression
model_lin_reg = LinearRegression()
```

Metriks yang akan dipakai adalah 'MAE(Mean Absolute Error)'

# 7 - Model Training

```python
model_lin_reg.fit(X_train_final,y_train)
```

# 8 - Model Evaluation

```python
y_pred_train = model_lin_reg.predict(X_train_final)
y_pred_test = model_lin_reg.predict(X_test_final)
y_pred_train
```

```python
y_pred_train.shape
```

```python
y_pred_test
```

```python
y_pred_test.shape
```

```python
from sklearn.metrics import mean_absolute_error
mae_train = mean_absolute_error(y_train,y_pred_train)
mae_test  = mean_absolute_error(y_test,y_pred_test)
print('Error - Train :', mae_train)
print('Error - Test :', mae_test)
```
